```python
#ê¸°ë³¸ ëª¨ë“ˆ ë¶ˆëŸ¬ì˜¤ê¸°
import numpy as np
import numpy.linalg as lin
import matplotlib.pyplot as plt
import pandas as pd
import random
```

```python
#ì„¤ëª…ë³€ìˆ˜ x1, x2, x3
x1 = [95, 91, 66, 94, 68, 63, 12, 73, 93, 51, 13, 70, 63, 63, 97, 56, 67, 96, 75, 6]
x2 = [56, 27, 25, 1, 9, 80, 92, 69, 6, 25, 83, 82, 54, 97, 66, 93, 76, 59, 94, 9]
x3 = [57, 34, 9, 79, 4, 77, 100, 42, 6, 96, 61, 66, 9, 25, 84, 46, 16, 63, 53, 30]
```

```python
#   ì„¤ëª…ë³€ìˆ˜ë“¤ì„ í•˜ë‚˜ì˜ í–‰ë ¬ë¡œ
X = np.stack((x1,x2,x3),axis=0)
```

### 1-1) ë¨¼ì € PCAë¥¼ ì‹œì‘í•˜ê¸° ì „ ë°ì´í„° scaling!!!! ë§¤ìš° ì¤‘ìš”í•©ë‹ˆë‹¤!

```python
#scaling
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_std = scaler.fit_transform(X)  #sclaing í›„ ë°ì´í„°
```

```python
features = X_std.T
```

### 1-2) ë‹¤ìŒìœ¼ë¡œ ê³µë¶„ì‚° í–‰ë ¬ì„ êµ¬í•©ë‹ˆë‹¤

https://docs.scipy.org/doc/numpy/reference/generated/numpy.cov.html ë¥¼ ì°¸ê³ í•˜ì‹œë©´ ë„ì›€ì´ ë ê±°ì—ìš”

```python
cov_matrix = np.cov(features)
```

```python
cov_matrix
```

### 1-3) ì´ì œ ê³ ìœ ê°’ê³¼ ê³ ìœ ë²¡í„°ë¥¼ êµ¬í•©ë‹ˆë‹¤

ì‹¤ìŠµì½”ë“œë¥¼ ì°¸ê³ í•´ì£¼ì„¸ìš”

```python
eigenvalues = lin.eig(cov_matrix)[0]
eigenvectors = lin.eig(cov_matrix)[1]
```

```python
mat = np.zeros((3,3))
```

```python
mat[0][0] = eigenvalues[0]
mat[1][1] = eigenvalues[1]
mat[2][2] = eigenvalues[2]
```

### 1-5) ë§ˆì§€ë§‰ìœ¼ë¡œ ê³ ìœ  ë²¡í„° ì¶•ìœ¼ë¡œ ê°’ì„ ë³€í™˜í•©ë‹ˆë‹¤
numpy.concatenate

https://docs.scipy.org/doc/numpy/reference/generated/numpy.concatenate.html

```python
def new_coordinates(X,eigenvectors):
    for i in range(eigenvectors.shape[0]): 
      
        if i == 0:
          #ì²«ë²ˆì§¸ ë²¡í„°ë¥¼ ê¸°ì¤€ìœ¼ë¡œ í•´ë‹¹ spaceë¡œ íˆ¬ì˜í•´ ì°¨ì›ì¶•ì†Œ ìˆ˜í–‰
            new = [X.dot(eigenvectors.T[i])]
        else:
            new = np.concatenate((new,  [X.dot(eigenvectors.T[i])]  ),axis=0) 
            #ê·¸ ë‹¤ìŒ ë²¡í„°ë¥¼ ê¸°ì¤€ìœ¼ë¡œ í•´ë‹¹ spaceë¡œ íˆ¬ì˜í•´ ì°¨ì›ì¶•ì†Œ ìˆ˜í–‰
            #ë³€í™˜ëœ ë°ì´í„°ë¥¼ ê¸°ì¡´ì— ë³€í™˜ëœ ë°ì´í„° ëª©ë¡ì— ë¶™ì´ê¸°
    return new.T

# ëª¨ë“  ê³ ìœ  ë²¡í„° ì¶•ìœ¼ë¡œ ë°ì´í„°ë¥¼ projectioní•œ ê°’ì…ë‹ˆë‹¤
```

# 2) PCAë¥¼ êµ¬í˜„í•´ë³´ê¸°

ìœ„ì˜ ê³¼ì •ì„ í•˜ë‚˜ì˜ í•¨ìˆ˜ë¡œ í‘œí˜„í•´ì£¼ì„¸ìš”!

1)ì—ì„œ ì§„í–‰í•œ ê³¼ì •ì„ ê·¸ëŒ€ë¡œ ê°€ì ¸ì˜¤ë©´ ë©ë‹ˆë‹¤..! , 

ìœ„ì—ì„œ ë§Œë“  í•¨ìˆ˜(new_coordinate)ë„ ê°€ì ¸ì˜¤ì…”ì„œ ì‚¬ìš©í•´ë„ ì¢‹ì•„ìš”!

+í•˜ì§€ë§Œ ì´ í•¨ìˆ˜ëŠ” numberì´ë¼ëŠ” ì¸ìë¥¼ ë°›ì•„ì„œ pcaì—ì„œ ì£¼ì„±ë¶„ ê°œìˆ˜ë¥¼ ê³ ë ¤í•˜ë„ë¡ í•©ë‹ˆë‹¤!

ê³ ìœ ê°’ì´ í° ì¶• ìˆœì„œë¡œ ì •ë ¬í•˜ëŠ” ê³¼ì • + ì£¼ì„±ë¶„ ê°œìˆ˜ numberì„ ê³ ë ¤í•´ì•¼í•œë‹¤ëŠ” ì ì„ ìœ ì˜í•´ì„œ í•¨ìˆ˜ë¥¼ ì •ì˜í•´ì£¼ì„¸ìš”


```python
from sklearn.preprocessing import StandardScaler

def MYPCA(X,number):

  scaler = StandardScaler() #ë°ì´í„° í‘œì¤€í™” -> ë³€ìˆ˜ ìŠ¤ì¼€ì¼ ì°¨ì´ë¡œ ì¸í•œ ë¶„ì‚°ì˜ ì™œê³¡ ë°©ì§€
  X_std = scaler.fit_transform(X) 

  features=X_std.T #ë°ì´í„°ì— ëŒ€í•œ ê³µë¶„ì‚° í–‰ë ¬ ê³„ì‚°
  cov_matrix = np.cov(features)

  eigenvalues = lin.eig(cov_matrix)[0] #eigenvalues
  eigenvectors = lin.eig(cov_matrix)[1] #eigenvectors

  new_pca=new_coordinates(X_std,eigenvectors) #ëª¨ë“  ë²¡í„°ì— ëŒ€í•œ ì°¨ì› ì¶•ì†Œ ì§„í–‰

  sort_index=list(eigenvalues.argsort()[::-1]) 
  #eigenvectorë¥¼ ë‚´ë¦¼ì°¨ìˆœìœ¼ë¡œ ì •ë ¬í•˜ê³  ì›ë˜ eigenvector ë°°ì—´ì—ì„œ ê·¸ í•´ë‹¹ ê°’ì˜ ì¸ë±ìŠ¤ê°€ ë¬´ì—‡ì¸ì§€ íŒŒì•…
  #sort_indexì—ëŠ” eigenvalueì˜ í° ê°’ì˜ ìˆœì„œëŒ€ë¡œ indexê°€ ë‹´ê²¨ ìˆìŒ

  new=[] #numberì— ëª…ì‹œëœ ë§Œí¼ pcaì— ì£¼ì„±ë¶„ ê°¯ìˆ˜ë¥¼ ê³ ë ¤í•˜ì—¬ ìƒˆë¡œ ìƒì„±í•œ ë°ì´í„°
  for relation in range(number):
     new.append(list(new_pca[:,sort_index[relation]])) 
     #eigenvectorë¡œ ë³€í™˜ëœ ë°ì´í„° ì¤‘ì—ì„œ eigenvalueê°€ ë†’ì€ ìˆœì„œëŒ€ë¡œ ê°€ì ¸ì™€ì„œ ë¶™ì„

  return (np.array(new)).T
```

```python
MYPCA(X,3)

# ìƒˆë¡œìš´ ì¶•ìœ¼ë¡œ ì˜ ë³€í™˜ë˜ì–´ì„œ ë‚˜íƒ€ë‚˜ë‚˜ìš”?
# ìœ„ì—ì„œ í–ˆë˜ PCAë‘ì€ ì°¨ì´ê°€ ìˆì„ ìˆ˜ ìˆì–´ìš” 
# ì™œëƒí•˜ë©´ ìœ„ì—ì„œëŠ” ê³ ìœ ê°’ì´ í° ì¶• ìˆœì„œë¡œ ì •ë ¬ì„ ì•ˆí–ˆì—ˆê±°ë“ ìš”
```
