```
from random import *

#일반 쿠키 
#쿠키의 종류 상관없이 모두 필요한 것-> 쿠키 생성 (이름/체력/일반 진입 속도 특성 부여) & 이동 능력(이동 방향으로) & 공격당함(받은 공격력 크기 / 남은 체력 / (사망선고)) 
#Parameters 
#(1)상속 : (Cookie) : name & hp & speed & direction & damage
#Method 
#(1)상속 : (Cookie): __init__  & move & get_damage
class Cookie:
  def __init__(self,name,hp,speed): #쿠키 생성
    self.name=name #이름
    self.hp=hp #체력
    self.speed=speed #속도
    print("{0} 쿠키 생성\n".format(name))

  def move(self, direction): #이동 능력
     print("{0} : {1} 방향으로 이동 : [속도 {2}] \n".format(self.name,direction,self.speed)) #부여받은 방향(direction)과 쿠키의 속도(speed)로 이동

  def get_damage(self,damage): #공격 당하는 메소드 (damage : 받은 공격력)
    print("{0} 쿠키 : 공격받은 데미지 : {1} \n".format(self.name,damage)) #받은 공격력 명시
    self.hp-=damage
    print("{0} 쿠키 : 현재 체력 상태 : {1} \n".format(self.name,self.hp)) #체력 - 받은 공격력 = 남은 체력 명시
    if self.hp <=0:
      print("{0} 쿠키 : 사망 ".format(self.name)) #남은 체력이 0 이하면... 쿠키 사망....


#돌격형 쿠키
#돌격형 쿠키에게 필요한 것 -> 일반 쿠키(이름/체력/일반 진입 속도) + 공격력 특성 부여 & 공격 능력 (이동 방향으로, 자신이 가진 공격력만큼)
#Parameters 
#(1)상속 : (Cookie) : name & hp & speed & direction & damage
#(2)돌격형 쿠키 (TankCookie) : damage
#Method 
#(1)상속 : (Cookie): __init__  & move & get_damage
#(2)돌격형 쿠키(TankCookie) : make_damage
class TankCookie(Cookie): #돌격형 쿠키 (일반 쿠키 상속)
    def __init__(self,name,hp,speed,damage):
      Cookie.__init__(self,name,hp,speed) #일반 쿠키 특성 부여
      self.damage=damage #공격력 특성 부여
    
    def make_damage(self,direction): #공격 능력 부여 메소드 
      print("{0} 쿠키 : {1} 방향으로 적군 공격 : [공격력: {2}] \n".format(self.name,direction,self.damage))
      #해당 쿠키(name)은 부여받은 방향(direction)으로 자신이 가진 공격력(damage)만큼 공격


#구미호맛쿠키 - 돌격형 쿠키임
#Parameters 
#(1)상속 : (Cookie) : name & hp & speed & direction & damage
#(2)상속 : 돌격형 쿠키 (TankCookie) : damage
#Method
#(1)상속 : (Cookie): __init__  & move & get_damage
#(2)상속 : 돌격형 쿠키(TankCookie) : make_damage
#(3)구미호맛쿠키 스킬 : 여우볼 (hp를 사용하여 가동)
class 구미호맛(TankCookie):
  def __init__(self):
    TankCookie.__init__(self,"구미호맛",30,3,5)

  
  def 여우볼(self): #구미호맛쿠키 스킬 : 여우볼 :구미호에서 쿠키로 변신해 여우볼 날리고 다시 구미호로 돌아감
    if self.hp>10:
      self.hp-=8 #hp 8만큼을 소비해서 스킬 사용
      print("{0} 쿠키: 변신 [ 구미호 -> 쿠키 ] \n".format(self.name)) #주어진 쿠키(name)가 구미호에서 쿠키로 변신
      print("{0} 쿠키:  여우볼 발사 (HP 8 감소) \n".format(self.name)) #주어진 쿠키(name)가 여우볼 발사
      self.damage*=3 #돌격형 쿠키인 구미호맛 쿠키가 가진 공격력보다 3배 증가
      print("{0} 쿠키: 변신 [ 쿠키 -> 구미호 ] \n".format(self.name))
    else:
      print("{0} 쿠키: Hp 부족으로 재주넘기 사용 불가 \n".format(self.name)) #hp가 부족하여 주어진 쿠키(name) 여우볼 스킬 사용불가

#웨어울프맛 쿠키 - 돌격형 쿠키임
#Parameters 
#(1)상속 : (Cookie) : name & hp & speed & direction & damage
#(2)상속 : 돌격형 쿠키 (TankCookie) : damage
#(3)웨어울프맛 : 각성 (각성 기능을 가동해 울프웨어맛 쿠키의 스킬 사용)
#Method
#(1)상속 : (Cookie): __init__  & move & get_damage
#(2)상속 : 돌격형 쿠키(TankCookie) : make_damage
#(3)웨어울프맛쿠키 스킬 : control_각성 (각성을 켜고 끄는 기능)
class 웨어울프맛(TankCookie):
  skill_developed=False

  def __init__(self):
    TankCookie.__init__(self,"웨어울프맛",40,10,6)
    self.각성=False #각성 기능 추가

  def control_각성(self): #웨어울프맛쿠키 스킬 : 각성 : 각성 기능을 켜서 사용함
    if 웨어울프맛.skill_developed==False:
      return

    
    if self.각성==False: #현재 각성 아님
      print("{0} 쿠키: 변신 [ 쿠키 -> 늑대 ] \n".format(self.name)) #주어진 쿠키(name)가 웨어울프맛쿠키가 늑대로 변신
      print("{0} 쿠키: 각성 상태로 전환 \n".format(self.name)) #각성 기능 킴
      self.damage*=2 #공격 2배로
      self.damage*=2 #공격 2배로
      self.seize_mode=True #각성 기능 켠 상태로 전환

     
    else: #현재 각성 켠 상태임
      print("{0} 쿠키: 각성 상태 해제 \n".format(self.name)) ##주어진 쿠키(name)가 각성 기능 끔 공지
      self.damage/=2
      self.각성=False #각성 기능 끈 상태로 전환


#자색고구마맛 쿠키 - 돌격형 쿠키임
#Parameters 
#(1)상속 : (Cookie) : name & hp & speed & direction & damage
#(2)상속 : 돌격형 쿠키 (TankCookie) : damage
#(3)자색고구마맛 : 자색토네이도 (자색토네이도 기능을 가동해 자색고구마맛 쿠키의 스킬 사용)
#Method
#(1)상속 : (Cookie): __init__  & move & get_damage
#(2)상속 : 돌격형 쿠키(TankCookie) : make_damage
#(3)자색고구마맛쿠키 스킬 : control_자색토네이도 (자색토네이도를 켜고 끄는 기능)
class 자색고구마맛(TankCookie):
  skill_developed=False

  def __init__(self):
    TankCookie.__init__(self,"자색고구마맛",20,10,6)
    self.자색토네이도=False #자색 토네이도 기능 추가

  def control_자색토네이도(self): #자색고구마맛쿠키 스킬 : 자색토네이도 : 자색토네이도 기능을 켜서 사용함
    if 자색고구마맛.skill_developed==False:
      return

    #현재 자색 토네이도 아님 -> 자색 토네이도 상태로 전환
    if self.자색토네이도==False: #현재 자색토네이도 아님
      print("{0} 쿠키: 자색토네이도 발사 \n".format(self.name)) #주어진 쿠키(name)가 자색토네이도를 발사함
      print("{0} 쿠키: 고구마 곤봉 날리기 \n".format(self.name)) #주어진 쿠키(name)가 고구마 곤봉을 날림
      self.damage*=2 #공격력 증가
      print("{0} 쿠키: 고구마 곤봉 날리기 \n".format(self.name)) #주어진 쿠키(name)가 고구마 곤봉을 날림
      self.damage*=2 #공격력 증가 =>고구마 곤봉 2연타!!!
      self.seize_mode=True  #자색 토네이도 기능 켠 상태로 전환


    else: #현재 자색토네이도 켠 상태임
      print("{0} 쿠키: 자색 토네이도 상태 해제 \n".format(self.name))
      self.damage/=2
      self.자색토네이도=False #자색토네이도 기능 끈 상태로 전환



#잠입형 쿠키
#parameters
#(1)infilt_speed : 잠입 스티드
#method
#(1)infiltration : 잠입하러가는 기능 
class infiltrateCookie:
  def __init__(self,infilt_speed): 
    self.infilt_speed=infilt_speed #잠입 스피드 부여

  def infiltration(self,name,direction):
    print("{0} 쿠키: {1} 방향으로 침투 [속도 {2}] \n".format(name,direction,self.infilt_speed))
    #주어진 쿠키(name)가 주어진 방향(directon)으로 해당 쿠키가 가진 잠입 스피드(infilt_speed)만큼 침투

#잠입형 공격 쿠키 (infiltrate_attack_Cookie) - 일반적으로 전진해서 공격하는 그냥 공격형 쿠키(TankCookie)와 다름 (이 쿠키는 잠입해서!!! 공격!!!)
#parameters
#(1)상속 : (Cookie) : name & hp & speed & direction & damage
#(2)상속 : (TankCookie) : damage
#(3)상속 : infiltrateCookie : infilt_speed 
#다중 상속 : TankCookie + infiltrateCookie
#Method
#(1)상속 : (Cookie): __init__  & move & get_damage
#(2)상속 : (TankCookie) : make_damage
#(3)상속 : (infiltrateCookie) : infiltration -> 이용해서 move 기능 실형
class infiltrate_attack_Cookie(TankCookie,infiltrateCookie):
  def __init__(self,name,hp,damage,infilt_speed):
    TankCookie.__init__(self,name,hp,0,damage) #일반 공격 전진 스피드는 0
    infiltrateCookie.__init__(self,infilt_speed)  #잠입 스피드 따로 부여

  def move(self, direction):
    self.infiltration(self.name,direction) #잠입하기

#뱀파이어맛 쿠키 - 잠입형 공격 쿠키
#parameters
#(1)상속 : (Cookie) : name & hp & speed & direction & damage
#(2)상속 : (TankCookie) : damage
#(3)상속 : (infiltrateCookie) : infilt_speed   => 다중 상속 : TankCookie + infiltrateCookie
#(4)뱀파이어맛 : blood - 흡혈모드 (흡혈 기능을 사용해 뱀파이어맛 쿠키 기능 구현)
#Method
#(1)상속 : (Cookie): __init__  & move & get_damage
#(2)상속 : (TankCookie) : make_damage
#(3)상속 : (infiltrateCookie) : infiltration -> 이용해서 move 기능 실형
#(4)뱀파이어맛쿠키 스킬 : drink_blood (흡혈모드를 켜고 끄는 기능)
class 뱀파이어맛(infiltrate_attack_Cookie):
  def __init__(self):
    infiltrate_attack_Cookie.__init__(self,"뱀파이어맛",90,30,10)
    self.blood=False #흡혈 모드(해제 상태)

  def drink_blood(self):
    if self.blood == True: #흡혈 모드 켜짐 -> 모드 해제
      print("{0} 쿠키: 흡혈 모드 해제 \n".format(self.name))
      self.blood=False
    else: #흡혈 모드 꺼짐 -> 흡혈모드 켬
      print("{0} 쿠키: 흡혈 모드 가동 \n".format(self.name))
      self.blood=True


#지원형 쿠키
#지원형 쿠키에게 필요한 것 -> 일반 쿠키(이름/체력/일반 진입 속도) + 지원 특성 부여(적군의 공격 정지) 
#Parameters 
#(1)상속 : (Cookie) : name & hp & speed & direction & damage
#Method 
#(1)상속 : (Cookie): __init__  & move & get_damage
#(2)지원형 쿠키(SupportCookie) : make_support
class SupportCookie(Cookie): #지원형 쿠키 -> 적군의 공격 정지 시킴
    def __init__(self,name,hp,speed):
      Cookie.__init__(self,name,hp,speed)
    
    def make_support(self):
      print("{0} 쿠키 : 적군 공격 정지  \n".format(self.name))

#민트초코맛쿠키 - 지원형 쿠키임
#Parameters 
#(1)상속 : (Cookie) : name & hp & speed & direction & damage
#Method
#(1)상속 : (Cookie): __init__  & move & get_damage
#(2)상속 : 지원형 쿠키(SupportCookie) : make_support
#(3)민트초코맛쿠키 스킬 : 바이올린연주 (바이올린 연주 실행해서 지원 기능 실행) 
class 민트초코맛(SupportCookie):
  def __init__(self):
    SupportCookie.__init__(self,"민트초코맛",30,6)

  
  def 바이올린연주(self):
    print("{0} 쿠키 : 바이올린 연주 시작  \n".format(self.name))
    self.make_support()

def game_start():
  print("[알림] 새로운 게임 시작 \n")

def game_over():
  print("Player : GG ")
  print("[Player]님 게임에서 퇴장 \n")

game_start()


구미호맛쿠키1=구미호맛()


웨어울프맛쿠키1=웨어울프맛()
웨어울프맛쿠키2=웨어울프맛()


자색고구마맛쿠키1=자색고구마맛()

뱀파이어맛쿠키1=뱀파이어맛()
뱀파이어맛쿠키2=뱀파이어맛()

민트초코맛쿠키1=민트초코맛()

#공격팀 일괄 관리
team=[]
team.append(구미호맛쿠키1)


team.append(웨어울프맛쿠키1)
team.append(웨어울프맛쿠키2)


team.append(자색고구마맛쿠키1)

team.append(뱀파이어맛쿠키1)
team.append(뱀파이어맛쿠키2)

#전군 이동
for cookie in team:
  cookie.move("1시")

#스킬 모드 개발
웨어울프맛.skill_developed=True
print("[알림] : 웨어울프맛 쿠키 각성 모드 개발 완료 \n")
자색고구마맛.skill_developed=True
print("[알림] : 자색고구마맛 쿠키 자색 토네이도 모드 개발 완료 \n")

for cookie in team:
  if isinstance(cookie,구미호맛):
    cookie.여우볼()
  elif isinstance(cookie,웨어울프맛):
    cookie.control_각성()
  elif isinstance(cookie,자색고구마맛):
    cookie.control_자색토네이도()
  elif isinstance(cookie,뱀파이어맛):
    cookie.drink_blood()


for cookie in team: #공격!!!
  cookie.make_damage("1시")

for cookie in team: #공격 받음...
  cookie.get_damage(randint(5,18))

민트초코맛쿠키1.바이올린연주() #지원

for cookie in team: #공격!!
  cookie.make_damage("4시")

game_over() #걍 끝냄
```

* `SRP` : 동일한 이유로 변경되는 것은 함께 모아져 있고 서로 다른 이유로 변경되는 것은 분리 되어 있기 때문 
* `OCP` : 쿠키에 할당된 기능과 정보를 변경하고 확장 할 수 있지만 그 기능을 사용하는 코드는 수정되지 않기 때문
* `LSP` : 자식 클래스는 부모 클래스가 따르던 계약 사항을 자식도 따르가 때문 부모 클래스를 사용하는 기존의 코드가 자식 클래스로 대입 시켰을 때 문제 없이 작동함
* `ISP` : 클라이언트가 사용하지 않는 매서드에 의존하지 않음. 즉, 자신과 관련없는 매서드는 구현되어 있지 않음
* `DIP` : 상위 모듈은 하위 모듈에 의존하지 않고 둘 다 추상화에 의존함. 또 상속 관계를 맺을 때 자주 변화하는 특정한 스킬을 가진 쿠키들에 의존하는 것이 아니라 공격형, 일반 쿠키와 같은 변화하기 어려운 거의 변화가 없는 것에 의존함.
